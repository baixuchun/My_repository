## 机器学习
* SVD解释的总方差约为23.99%。这表示降维后的两个主要成分能够解释原始数据约24%的方差。

```python
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 从文件中加载数据
data = pd.read_csv('catalyst_dataset.csv')

# 数据标准化
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

# 应用SVD
n_components = 2  # 选择降维后的特征数
svd = TruncatedSVD(n_components=n_components)
svd.fit(scaled_data)
reduced_data = svd.transform(scaled_data)

# 可视化SVD的结果
plt.figure(figsize=(10, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.title('SVD Reduced Catalyst Dataset')
plt.xlabel('SVD Feature 1')
plt.ylabel('SVD Feature 2')
plt.grid(True)
plt.show()

# 输出SVD的一些关键信息
explained_variance = svd.explained_variance_ratio_.sum()
explained_variance
```
