```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 设置随机种子以保证结果可复现
np.random.seed(42)

# 生成模拟数据
cluster_1 = np.random.normal(loc=[5, 5], scale=[0.5, 0.5], size=(50, 2))  # 簇1
cluster_2 = np.random.normal(loc=[10, 10], scale=[1.0, 1.0], size=(70, 2))  # 簇2
cluster_3 = np.random.normal(loc=[15, 5], scale=[0.8, 0.8], size=(60, 2))  # 簇3

# 生成一些噪声点
noise = np.random.uniform(low=0, high=20, size=(10, 2))

# 合并所有点
data = np.vstack([cluster_1, cluster_2, cluster_3, noise])

# 生成颗粒的大小和形状特征
particle_size = np.random.uniform(0.1, 2.0, size=len(data))  # 颗粒大小
particle_shape = np.random.uniform(0.5, 1.5, size=len(data))  # 颗粒形状

# 创建 DataFrame 保存数据
df = pd.DataFrame(data, columns=['x', 'y'])
df['size'] = particle_size
df['shape'] = particle_shape

# 将数据保存为 CSV 文件
csv_filename = 'polymer_microstructure_data.csv'
df.to_csv(csv_filename, index=False)

print(f'Data saved to {csv_filename}')

# 可视化生成的数据点
plt.figure(figsize=(8, 6))
plt.scatter(df['x'], df['y'], c='blue', s=df['size'] * 20, alpha=0.6, label='Particles')
plt.title('Generated Data for Polymer Material Analysis')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```
