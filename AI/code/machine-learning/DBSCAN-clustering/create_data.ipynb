```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 
np.random.seed(42)

# 
cluster_1 = np.random.normal(loc=[5, 5], scale=[0.5, 0.5], size=(50, 2))  # 1
cluster_2 = np.random.normal(loc=[10, 10], scale=[1.0, 1.0], size=(70, 2))  # 2
cluster_3 = np.random.normal(loc=[15, 5], scale=[0.8, 0.8], size=(60, 2))  # 3

# 
noise = np.random.uniform(low=0, high=20, size=(10, 2))

# 
data = np.vstack([cluster_1, cluster_2, cluster_3, noise])

# 
particle_size = np.random.uniform(0.1, 2.0, size=len(data))  # 
particle_shape = np.random.uniform(0.5, 1.5, size=len(data))  # 

#  DataFrame 
df = pd.DataFrame(data, columns=['x', 'y'])
df['size'] = particle_size
df['shape'] = particle_shape

#  CSV 
csv_filename = 'polymer_microstructure_data.csv'
df.to_csv(csv_filename, index=False)

print(f'Data saved to {csv_filename}')

# 
plt.figure(figsize=(8, 6))
plt.scatter(df['x'], df['y'], c='blue', s=df['size'] * 20, alpha=0.6, label='Particles')
plt.title('Generated Data for Polymer Material Analysis')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.legend()
plt.grid(True)
plt.show()
```
