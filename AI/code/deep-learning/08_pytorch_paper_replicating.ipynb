## 什么是机器学习研究论文？

机器学习研究论文是一种科学论文，详细记录了研究团队在特定领域的研究成果。

虽然不同论文的内容可能有所不同，但它们通常遵循以下结构：

| **部分** | **内容** |
| ----- | ----- |
| **摘要 (Abstract)** | 论文主要发现和贡献的概述/总结。 |
| **引言 (Introduction)** | 论文的主要问题是什么，以及之前尝试解决该问题的方法的详细信息。 |
| **方法 (Method)** | 研究人员是如何进行研究的？例如，使用了哪些模型、数据来源和训练设置？ |
| **结果 (Results)** | 论文的研究结果是什么？如果使用了新模型或训练设置，这些结果与之前的工作相比如何？（此处**实验追踪**非常有用） |
| **结论 (Conclusion)** | 所提出方法的局限性是什么？研究界的下一步是什么？ |
| **参考文献 (References)** | 研究人员参考了哪些资源/其他论文来构建他们的研究工作？ |
| **附录 (Appendix)** | 是否有未包含在上述部分中的额外资源/研究发现？ |

具体来说，我们将使用 PyTorch 复现机器学习研究论文 [*An Image is Worth 16x16 Words: Transformers for Image Recognition at Scale*](https://arxiv.org/abs/2010.11929)（ViT 论文）。

Transformer 神经网络架构最初在论文 [*Attention is all you need*](https://arxiv.org/abs/1706.03762) 中提出。

最初的 Transformer 架构是为一维（1D）文本序列设计的。

**Transformer 架构**通常被认为是使用[**注意力机制**](https://en.wikipedia.org/wiki/Attention_(machine_learning))作为主要学习层的任何神经网络，类似于卷积神经网络（CNN）以卷积作为主要学习层。

顾名思义，**Vision Transformer (ViT) 架构旨在将原始的 Transformer 架构适配到视觉问题上**（分类是第一个目标，随后扩展到了更多任务）。

尽管原始的 Vision Transformer 在过去几年中经历了多次迭代，但我们将重点复现其最初的版本，也被称为“基础版 Vision Transformer”。因为一旦掌握了原始版本的实现，你就能轻松适应其他版本。

### 2.2 将图像转换为 `DataLoader`

变换已创建！

现在让我们生成 `DataLoader`。

ViT 论文提到使用的批量大小为 4096，这相当于我们之前使用的批量大小（32）的 128 倍。

然而，我们将坚持使用批量大小为 32。

为什么？

因为某些硬件（包括 Google Colab 的免费版本）可能无法处理批量大小为 4096。

批量大小为 4096 意味着一次需要将 4096 张图像加载到 GPU 内存中。

当你有像 Google 的研究团队那样的强大硬件时，这样的设置是可行的，但如果你运行在单个 GPU（例如使用 Google Colab）上，先确保用较小的批量大小正常运行是一个好主意。

### 3.1 输入与输出、层与块

ViT 是一种深度学习神经网络架构。

任何神经网络架构通常由 **层（layer）** 组成。

而多层的集合通常称为 **块（block）**。

将多个块堆叠在一起，就构成了完整的架构。

一个 **层** 接收一个输入（例如图像张量），对其执行某种函数操作（例如层中的 `forward()` 方法），然后返回一个输出。

因此，如果一个 **单层** 接收输入并返回输出，那么多个层的集合或一个 **块** 也会接收输入并返回输出。

让我们具体说明：
- **层（Layer）** - 接收输入，对其执行某种函数操作，返回输出。
- **块（Block）** - 多个层的集合，接收输入，对其执行一系列函数操作，返回输出。
- **架构（或模型）（Architecture or model）** - 多个块的集合，接收输入，对其执行一系列函数操作，返回输出。

我们将基于这一理念复现 ViT 论文。

我们将逐层、逐块、逐函数地拼接这些部分，就像搭积木一样，最终构建出我们期望的整体架构。

这样做的原因是，直接阅读整篇研究论文可能会让人感到难以入手。

为了更好地理解，我们会将其拆解，从单层的输入和输出开始，一步步延伸到整个模型的输入和输出。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-intputs-outputs-layers-and-blocks.png" alt="模型的输入和输出、层和块" width=900/>

*现代深度学习架构通常由层和块组成。层接收输入（以数值表示的数据），通过某种函数操作处理数据（例如上图中的自注意力公式，但这一函数可以是几乎任何操作），然后输出数据。块通常是多个层的堆叠，与单层的操作类似，但会重复多次。*

#### 3.2.1 探索图 1

让我们从 ViT 论文的图 1 开始。

我们需要关注的主要内容是：
1. **层（Layers）** - 接收一个**输入**，对其执行操作或函数，产生一个**输出**。
2. **块（Blocks）** - 层的集合，同样接收一个**输入**并产生一个**输出**。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-figure-1-inputs-and-outputs.png" width=900 alt="ViT 论文的图 1，显示输入、输出、层和块"/>

*ViT 论文的图 1，展示了构成架构的不同输入、输出、层和块。我们的目标是使用 PyTorch 代码复现每一个组件。*

ViT 架构由以下几个阶段组成：
- **Patch + Position Embedding（输入）** - 将输入图像分割为一系列图像块（patch），并添加一个位置编号以指定块的顺序。
- **Linear projection of flattened patches（嵌入块）** - 图像块被转换为一个**嵌入向量**。使用嵌入的好处是，它是图像的一种*可学习*表示（通常以向量形式），可以随着训练而改进。
- **Norm** - 是 "[Layer Normalization](https://paperswithcode.com/method/layer-normalization)" 或 "LayerNorm" 的缩写，一种用于正则化（减少过拟合）神经网络的技术。可以通过 PyTorch 层 [`torch.nn.LayerNorm()`](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html) 使用。
- **Multi-Head Attention** - 即 [多头自注意力层](https://paperswithcode.com/method/multi-head-attention)（简称 "MSA"）。可以通过 PyTorch 层 [`torch.nn.MultiheadAttention()`](https://pytorch.org/docs/stable/generated/torch.nn.MultiheadAttention.html) 创建。
- **MLP（或 [多层感知机](https://en.wikipedia.org/wiki/Multilayer_perceptron)）** - 通常指任何前馈层的集合（在 PyTorch 中是具有 `forward()` 方法的层集合）。在 ViT 论文中，作者将 MLP 称为 "MLP block"，包含两个 [`torch.nn.Linear()`](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html) 层，中间使用 [`torch.nn.GELU()`](https://pytorch.org/docs/stable/generated/torch.nn.GELU.html) 非线性激活函数（见第 3.1 节），并在每个层后添加 [`torch.nn.Dropout()`](https://pytorch.org/docs/stable/generated/torch.nn.Dropout.html) 层（见附录 B.1）。
- **Transformer Encoder** - Transformer 编码器是上述各层的集合。编码器中有两个跳跃连接（"+" 符号），即输入直接传递给临近层和后续层。整个 ViT 架构由多个堆叠的 Transformer 编码器组成。
- **MLP Head** - 这是架构的输出层，将输入的学习特征转换为类别输出。对于图像分类问题，可以称其为 "分类头"。MLP Head 的结构与 MLP block 类似。

你可能会注意到，ViT 架构中的许多组件可以通过现有的 PyTorch 层创建。

这正是 PyTorch 的设计初衷之一，为研究人员和机器学习从业者提供可重用的神经网络层。

#### 3.2.2 探索四个公式

接下来，我们将关注 ViT 论文第 3.1 节中的四个公式。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-four-equations.png" width=650 alt="ViT 机器学习论文中的四个数学公式"/>

*这四个公式描述了 ViT 架构的四个主要部分的数学原理。*

第 3.1 节详细说明了这些公式（为简洁起见省略了一部分文本，加粗部分是重点）：

| **公式编号** | **ViT 论文第 3.1 节中的描述** |
| ----- | ----- |
| 1 | ...Transformer 在所有层中使用固定的潜在向量大小 $D$，因此我们将图像块展开并映射到 $D$ 维，通过一个**可训练的线性投影**（公式 1）。我们将这种投影的输出称为**块嵌入（patch embeddings）**...**位置嵌入（position embeddings）** 被添加到块嵌入中以保留位置信息。我们使用标准的**可学习一维位置嵌入**...|

让我们将这些描述映射到图 1 中的 ViT 架构。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-mapping-the-four-equations-to-figure-1.png" width=1000 alt="将 ViT 论文的图 1 映射到文中描述的四个公式"/>

*将 ViT 论文中的图 1 与第 3.1 节中的四个公式联系起来，解释每个层/块背后的数学原理。*

上图包含了许多内容，但通过跟随彩色线条和箭头，可以揭示 ViT 架构的主要概念。

我们进一步分解每个公式（我们的目标是用代码复现这些公式），如何？

在所有公式中（除公式 4 之外），"$\mathbf{z}$" 是某一层的原始输出：
1. $\mathbf{z}_{0}$ 是 "z 零"（这是初始块嵌入层的输出）。
2. $\mathbf{z}_{\ell}^{\prime}$ 是 "z 某层的中间值"。
3. $\mathbf{z}_{\ell}$ 是 "z 某层的输出"。

而 $\mathbf{y}$ 是整个架构的最终输出。

#### 3.2.3 公式 1 概述

$$
\begin{aligned}
\mathbf{z}_{0} &=\left[\mathbf{x}_{\text {class }} ; \mathbf{x}_{p}^{1} \mathbf{E} ; \mathbf{x}_{p}^{2} \mathbf{E} ; \cdots ; \mathbf{x}_{p}^{N} \mathbf{E}\right]+\mathbf{E}_{\text {pos }}, & & \mathbf{E} \in \mathbb{R}^{\left(P^{2} \cdot C\right) \times D}, \mathbf{E}_{\text {pos }} \in \mathbb{R}^{(N+1) \times D}
\end{aligned}
$$

该公式处理输入图像的分类标记（class token）、块嵌入（patch embedding）和位置嵌入（position embedding）（$\mathbf{E}$ 表示嵌入）。

在向量形式中，嵌入可能看起来如下：

```python
x_input = [class_token, image_patch_1, image_patch_2, image_patch_3...] + [class_token_position, image_patch_1_position, image_patch_2_position, image_patch_3_position...]
```

其中向量中的每个元素都是可学习的（它们的 `requires_grad=True`）。

#### 3.2.4 公式 2 概述

$$
\begin{aligned}
\mathbf{z}_{\ell}^{\prime} &=\operatorname{MSA}\left(\operatorname{LN}\left(\mathbf{z}_{\ell-1}\right)\right)+\mathbf{z}_{\ell-1}, & & \ell=1 \ldots L
\end{aligned}
$$

这表示从第 $1$ 层到第 $L$ 层（总层数）都有一个多头自注意力层（MSA），包裹着一个 LayerNorm 层（LN）。

最后的加法等价于将输入添加到输出，形成了一个[跳跃/残差连接](https://paperswithcode.com/method/residual-connection)。

我们将这一层称为“MSA 块”。

伪代码可能如下所示：

```python
x_output_MSA_block = MSA_layer(LN_layer(x_input)) + x_input
```

注意最后的跳跃连接（将层的输入添加到层的输出）。

#### 3.2.5 公式 3 概述

$$
\begin{aligned}
\mathbf{z}_{\ell} &=\operatorname{MLP}\left(\operatorname{LN}\left(\mathbf{z}_{\ell}^{\prime}\right)\right)+\mathbf{z}_{\ell}^{\prime}, & & \ell=1 \ldots L \\
\end{aligned}
$$

这表示从第 $1$ 层到第 $L$ 层（总层数）都有一个多层感知机层（MLP），包裹着一个 LayerNorm 层（LN）。

最后的加法表示存在一个跳跃/残差连接。

我们将这一层称为“MLP 块”。

伪代码可能如下所示：

```python
x_output_MLP_block = MLP_layer(LN_layer(x_output_MSA_block)) + x_output_MSA_block
```

注意最后的跳跃连接（将层的输入添加到层的输出）。

#### 3.2.6 公式 4 概述

$$
\begin{aligned}
\mathbf{y} &=\operatorname{LN}\left(\mathbf{z}_{L}^{0}\right) & &
\end{aligned}
$$

这表示对于最后一层 $L$，输出 $y$ 是 $z$ 的索引 0 位置的标记，经过一个 LayerNorm 层（LN）的处理。

在我们的情况下，就是 `x_output_MLP_block` 的索引 0 位置：

```python
y = Linear_layer(LN_layer(x_output_MLP_block[0]))
```

当然，上述表达有所简化，我们将在为每个部分编写 PyTorch 代码时处理这些细节。

## 4. 公式 1：将数据分割为块并创建类别、位置和块嵌入

我记得有位机器学习工程师朋友曾说过：“关键在于嵌入。”

意思是，如果你能以一种良好的、可学习的方式表示你的数据（**嵌入是可学习的表示**），那么学习算法很可能会在这些数据上表现良好。

基于这一点，我们从为 ViT 架构创建类别、位置和块嵌入开始。

我们先从**块嵌入（patch embedding）**着手。

这意味着我们将输入图像转化为一系列的图像块，然后对这些块进行嵌入。

回想一下，**嵌入**是一种可学习的表示形式，通常是一个向量。

“可学习”这一术语非常重要，因为它意味着输入图像的数值表示（即模型接收到的数据表示）可以随着时间不断优化。

我们将遵循 ViT 论文第 3.1 节开头的一段话（加粗为重点）：

> 标准 Transformer 接收的输入是一维 token 嵌入序列。为了处理二维图像，我们将图像 $\mathbf{x} \in \mathbb{R}^{H \times W \times C}$ 重塑为一系列展平的二维图像块 $\mathbf{x}_{p} \in \mathbb{R}^{N \times\left(P^{2} \cdot C\right)}$，其中 $(H, W)$ 是原始图像的分辨率，$C$ 是通道数，$(P, P)$ 是每个图像块的分辨率，$N=H W / P^{2}$ 是生成的块数，同时也是 Transformer 的有效输入序列长度。Transformer 在所有层中使用固定的潜在向量大小 $D$，因此我们将图像块展平并映射到 $D$ 维，通过一个可训练的线性投影（公式 1）。我们将这种投影的输出称为**块嵌入（patch embeddings）**。

考虑到我们正在处理图像形状，让我们记住 ViT 论文表 3 中的一行内容：

> 训练分辨率为 **224**。

让我们分解上述文本的内容。

- $D$ 是**块嵌入**的大小，不同大小的 ViT 模型的 $D$ 值可在表 1 中找到。
- 图像最初是二维的，大小为 ${H \times W \times C}$。
  - $(H, W)$ 是原始图像的分辨率（高度和宽度）。
  - $C$ 是通道数。
- 图像被转换为一系列展平的二维图像块，大小为 ${N \times\left(P^{2} \cdot C\right)}$。
  - $(P, P)$ 是每个图像块的分辨率（**块大小**）。
  - $N=H W / P^{2}$ 是生成的块数，同时也是 Transformer 的输入序列长度。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-equation-1-annotated.png" width=900 alt="将 ViT 架构中的位置和块嵌入部分映射到描述其数学公式的相关内容"/>

*将 ViT 架构图中位置和块嵌入部分与公式 1 进行映射。第 3.1 节开头描述了块嵌入层的不同输入和输出形状。*

### 4.1 手动计算块嵌入的输入和输出形状

我们先通过手动计算这些输入和输出形状值来开始吧。

为此，让我们创建一些变量来模拟上文中的各个术语（例如 $H$、$W$ 等）。

我们将使用块大小（$P$）为 16，因为这是表现最好的 ViT-Base 版本所使用的值（详见 ViT 论文表 5 中 "ViT-B/16" 列）。

````python
# 创建示例值
height = 224 # H ("训练分辨率为 224")
width = 224 # W
color_channels = 3 # C
patch_size = 16 # P

# 计算 N（块的数量）
number_of_patches = int((height * width) / patch_size**2)
print(f"Number of patches (N) with image height (H={height}), width (W={width}) and patch size (P={patch_size}): {number_of_patches}")
````

我们已经得到了块的数量，那么我们也来创建块嵌入层的图像输出大小吧？

更进一步，让我们复现块嵌入层的输入和输出形状。

回顾：

- **输入：** 图像最初是二维的，大小为 ${H \times W \times C}$。
- **输出：** 图像被转换为一系列展平的二维图像块，大小为 ${N \times\left(P^{2} \cdot C\right)}$。

````python
# 输入形状（这是单张图片的大小）
embedding_layer_input_shape = (height, width, color_channels)

# 输出形状
embedding_layer_output_shape = (number_of_patches, patch_size**2 * color_channels)

print(f"Input shape (single 2D image): {embedding_layer_input_shape}")
print(f"Output shape (single 2D image flattened into patches): {embedding_layer_output_shape}")
````

输入和输出形状已获取！

### 4.2 将单张图片转化为补丁（patches）

现在我们知道了**补丁嵌入**（**patch embedding**）层的理想输入和输出形状，接下来我们开始构建它。

我们的做法是将整体架构拆解成更小的部分，重点关注每个层的输入和输出。

那么，如何创建补丁嵌入层呢？

我们马上就会讲到这个，首先，让我们*可视化，可视化，再可视化！* 了解将图片转化为补丁的过程。

我们从单张图片开始。

````python
# 查看单张图片
plt.imshow(image.permute(1, 2, 0)) # 调整以适应matplotlib
plt.title(class_names[label])
plt.axis(False);
````

我们想将这张图片转化为与ViT论文中的图1一致的补丁。

我们先从可视化补丁化后的像素的最上面一行开始怎么样？

我们可以通过在不同的图像维度上进行索引来实现这一点。

````python
# 改变图像形状以兼容matplotlib (color_channels, height, width) -> (height, width, color_channels)
image_permuted = image.permute(1, 2, 0)

# 索引以绘制补丁化后的像素的最上面一行
patch_size = 16
plt.figure(figsize=(patch_size, patch_size))
plt.imshow(image_permuted[:patch_size, :, :]);
````

现在我们已经得到了最上面一行，接下来将它转化为补丁。

我们可以通过迭代最上面一行中的补丁数量来实现这一点。

````python
# 设置超参数并确保img_size和patch_size兼容
img_size = 224
patch_size = 16
num_patches = img_size/patch_size
assert img_size % patch_size == 0, "Image size must be divisible by patch size"
print(f"Number of patches per row: {num_patches}\nPatch size: {patch_size} pixels x {patch_size} pixels")

# 创建一系列子图
fig, axs = plt.subplots(nrows=1,
                        ncols=img_size // patch_size, # one column for each patch
                        figsize=(num_patches, num_patches),
                        sharex=True,
                        sharey=True)

# 迭代最上面一行中的补丁数量
for i, patch in enumerate(range(0, img_size, patch_size)):
    axs[i].imshow(image_permuted[:patch_size, patch:patch+patch_size, :]); # keep height index constant, alter the width index
    axs[i].set_xlabel(i+1) # set the label
    axs[i].set_xticks([])
    axs[i].set_yticks([])
````

这些补丁看起来很不错！

那我们来对整个图像进行操作怎么样？

这次我们将迭代高度和宽度的索引，并将每个补丁作为一个独立的子图进行绘制。

````python
# 设置超参数并确保img_size和patch_size兼容
img_size = 224
patch_size = 16
num_patches = img_size/patch_size
assert img_size % patch_size == 0, "Image size must be divisible by patch size"
print(f"Number of patches per row: {num_patches}\
        \nNumber of patches per column: {num_patches}\
        \nTotal patches: {num_patches*num_patches}\
        \nPatch size: {patch_size} pixels x {patch_size} pixels")

# 创建一系列子图
fig, axs = plt.subplots(nrows=img_size // patch_size, # 需要整数而不是浮点数
                        ncols=img_size // patch_size,
                        figsize=(num_patches, num_patches),
                        sharex=True,
                        sharey=True)

# 遍历图像的高度和宽度
for i, patch_height in enumerate(range(0, img_size, patch_size)): # 遍历高度
    for j, patch_width in enumerate(range(0, img_size, patch_size)): # 遍历宽度

        # 绘制调整过的图像补丁 (image_permuted -> (高度, 宽度, 色彩通道))
        axs[i, j].imshow(image_permuted[patch_height:patch_height+patch_size, # 遍历高度
                                        patch_width:patch_width+patch_size, # 遍历宽度
                                        :]) # 获取所有色彩通道

        # 设置标签信息，移除坐标轴刻度以提高清晰度，并将标签设置为外部
        axs[i, j].set_ylabel(i+1,
                             rotation="horizontal",
                             horizontalalignment="right",
                             verticalalignment="center")
        axs[i, j].set_xlabel(j+1)
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])
        axs[i, j].label_outer()

# 设置总标题
fig.suptitle(f"{class_names[label]} -> Patchified", fontsize=16)
plt.show()
````

图像已补丁化！

看起来很酷。

### 4.3 使用 `torch.nn.Conv2d()` 创建图像补丁

我们已经看到了将图像转化为补丁的效果，现在我们开始使用PyTorch来复制补丁嵌入层。

为了可视化单张图像，我们编写了代码来遍历单张图像的不同高度和宽度维度，并绘制出各个补丁。

这个操作与我们在[03. PyTorch计算机视觉第7.1节：逐步了解`nn.Conv2d()`](https://www.learnpytorch.io/03_pytorch_computer_vision/#71-stepping-through-nnconv2d)中看到的卷积操作非常相似。

事实上，ViT论文的作者在第3.1节中提到，补丁嵌入是可以通过卷积神经网络（CNN）实现的：

> **混合架构**：作为原始图像补丁的替代，输入序列可以通过CNN的特征图（LeCun等，1989年）来形成。在这种混合模型中，补丁嵌入投影$\mathbf{E}$（公式1）应用于从**CNN特征图**中提取的补丁。作为特例，补丁可以具有空间大小$1 \times 1$，这意味着**输入序列通过简单地展平特征图的空间维度并投影到Transformer维度来获得**。分类输入嵌入和位置嵌入如上所述被添加。

他们所提到的“**特征图**”是卷积层在给定图像上进行卷积操作时产生的权重/激活值。

<img src="https://github.com/mrdbourke/pytorch-deep-learning/raw/main/images/08-vit-paper-patch-embedding-animation.gif" width=900 alt="通过卷积层对单张图像进行操作来创建补丁嵌入的示例"/>

*通过将[`torch.nn.Conv2d()`](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html)层的`kernel_size`和`stride`参数设置为`patch_size`，我们可以有效地获得一个将图像拆分为补丁并为每个补丁创建可学习嵌入（在ViT论文中称为“线性投影”）的层。*

还记得我们为补丁嵌入层设定的理想输入和输出形状吗？

* **输入：** 图像最初是2D的，大小为 ${H \times W \times C}$。
* **输出：** 图像被转换为大小为 ${N \times (P^{2} \cdot C)}$ 的一维序列，表示展平后的2D补丁。

例如，对于224大小的图像和16大小的补丁：

* **输入（二维图像）：** (224, 224, 3) -> (高度, 宽度, 色彩通道)
* **输出（展平后的2D补丁）：** (196, 768) -> (补丁数量, 嵌入维度)

我们可以使用以下方式重现这些：
* [`torch.nn.Conv2d()`](https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html) 用于将图像转换为CNN特征图的补丁。
* [`torch.nn.Flatten()`](https://pytorch.org/docs/stable/generated/torch.nn.Flatten.html) 用于展平特征图的空间维度。

让我们从 `torch.nn.Conv2d()` 层开始。

我们可以通过将 `kernel_size` 和 `stride` 设置为 `patch_size` 来复制补丁的创建过程。

这意味着每个卷积核的大小将是 `(patch_size x patch_size)`，或者如果 `patch_size=16`，则为 `(16 x 16)`（相当于一个完整的补丁）。

每一步或卷积核的 `stride` 将是 `patch_size` 像素长，或者 `16` 像素长（相当于步进到下一个补丁）。

我们将 `in_channels=3` 设置为图像的颜色通道数，`out_channels=768`，与ViT-Base表格1中的$D$值相同（这是嵌入维度，每张图像将被嵌入到一个大小为768的可学习向量中）。

````python
from torch import nn

# 设置补丁大小
patch_size = 16

# 创建Conv2d层，使用ViT论文中的超参数
conv2d = nn.Conv2d(in_channels=3, # 颜色通道数
                   out_channels=768, # 来自表1：隐藏大小D，这是嵌入大小
                   kernel_size=patch_size, # 也可以使用(patch_size, patch_size)
                   stride=patch_size,
                   padding=0)
````

Now we've got a convoluational layer, let's see what happens when we pass a single image through it.

````python
# 查看单张图片
plt.imshow(image.permute(1, 2, 0)) # 调整以适应matplotlib
plt.title(class_names[label])
plt.axis(False);
````

````python
# 将图像通过卷积层
image_out_of_conv = conv2d(image.unsqueeze(0)) # 添加一个批次维度 (height, width, color_channels) -> (batch, height, width, color_channels)
print(image_out_of_conv.shape)
````

通过将图像通过卷积层处理，它被转换成一系列768（这是嵌入大小或$D$）个特征/激活图。

因此，其输出形状可以解释为：

```python
torch.Size([1, 768, 14, 14]) -> [batch_size, embedding_dim, feature_map_height, feature_map_width]
```

接下来，我们来可视化五个随机的特征图，看看它们是什么样子的。

````python
# 绘制随机的5个卷积特征图
import random
random_indexes = random.sample(range(0, 758), k=5) # 从0到嵌入大小之间随机选择5个数字
print(f"Showing random convolutional feature maps from indexes: {random_indexes}")

# 创建图像
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(12, 12))

# 绘制随机图像特征图
for i, idx in enumerate(random_indexes):
    image_conv_feature_map = image_out_of_conv[:, idx, :, :] # 在卷积层的输出张量上进行索引
    axs[i].imshow(image_conv_feature_map.squeeze().detach().numpy())
    axs[i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[]);
````

注意到这些特征图在某种程度上都代表了原始图像，在可视化更多特征图之后，你可以开始看到不同的主要轮廓和一些主要特征。

需要注意的重要一点是，随着神经网络的学习，这些特征可能会随着时间发生变化。

正因为如此，这些特征图可以被视为我们图像的**可学习嵌入**（**learnable embedding**）。

接下来，让我们以数值的形式查看其中一个特征图。

````python
# 获取单个特征图的张量形式
single_feature_map = image_out_of_conv[:, 0, :, :]
single_feature_map, single_feature_map.requires_grad
````

`single_feature_map` 的 `grad_fn` 输出和 `requires_grad=True` 属性意味着 PyTorch 正在跟踪这个特征图的梯度，并且它将在训练过程中通过梯度下降进行更新。

### 4.4 使用 `torch.nn.Flatten()` 将补丁嵌入展平

我们已经将图像转化为补丁嵌入，但它们仍然是2D格式的。

如何将它们转换为ViT模型补丁嵌入层所需的输出形状呢？

* **所需输出（展平的2D补丁的1D序列）：** (196, 768) -> （补丁数，嵌入维度）-> ${N \times\left(P^{2} \cdot C\right)}$

我们来检查一下当前的形状。

````python
# 当前张量形状
print(f"Current tensor shape: {image_out_of_conv.shape} -> [batch, embedding_dim, feature_map_height, feature_map_width]")
````

好，我们已经得到了768部分（$(P^{2} \cdot C)$），但我们仍然需要补丁的数量（$N$）。

回顾ViT论文的第3.1节，它提到（加粗部分是我的）：

> 作为一个特例，补丁可以具有空间尺寸 $1 \times 1$，这意味着**输入序列是通过简单地*展平*特征图的空间维度并投影到Transformer维度**来获得的。

展平特征图的空间维度，对吧？

在PyTorch中，我们有什么层可以用来展平吗？

那就是[`torch.nn.Flatten()`](https://pytorch.org/docs/stable/generated/torch.nn.Flatten.html)。

但我们不想展平整个张量，我们只想展平“特征图的空间维度”。

在我们的例子中，这指的是 `image_out_of_conv` 中的 `feature_map_height` 和 `feature_map_width` 维度。

所以，如何创建一个 `torch.nn.Flatten()` 层，仅展平这两个维度呢？我们可以使用 `start_dim` 和 `end_dim` 参数来设置吗？

````python
# 创建展平层
flatten = nn.Flatten(start_dim=2, # 展平 feature_map_height（维度2）
                     end_dim=3) # 展平 feature_map_width（维度3）
````

现在让我们将这些步骤整合起来！

我们将：
1. 获取一张单独的图像。
2. 将其通过卷积层（`conv2d`），将图像转化为2D特征图（补丁嵌入）。
3. 将2D特征图展平成一个单一的序列。

````python
# 1. 查看单张图像
plt.imshow(image.permute(1, 2, 0)) # 调整为matplotlib兼容格式
plt.title(class_names[label])
plt.axis(False);
print(f"Original image shape: {image.shape}")

# 2. 将图像转化为特征图
image_out_of_conv = conv2d(image.unsqueeze(0)) # 添加批处理维度以避免形状错误
print(f"Image feature map shape: {image_out_of_conv.shape}")

# 3. 展平特征图
image_out_of_conv_flattened = flatten(image_out_of_conv)
print(f"Flattened image feature map shape: {image_out_of_conv_flattened.shape}")
````

哇哦！看起来我们的 `image_out_of_conv_flattened` 形状非常接近我们所需的输出形状：

* **所需输出（展平的2D补丁）：** (196, 768) -> ${N \times\left(P^{2} \cdot C\right)}$
* **当前形状：** (1, 768, 196)

唯一的区别是当前形状有一个批处理维度，而且维度的顺序与所需输出不同。

我们该如何修复这个问题呢？

好吧，如何重新排列这些维度呢？

我们可以像在使用matplotlib绘制图像时重新排列图像张量一样，使用 `torch.Tensor.permute()` 来实现。

我们试试看。

````python
# 获取正确形状的展平图像补丁嵌入
image_out_of_conv_flattened_reshaped = image_out_of_conv_flattened.permute(0, 2, 1) # [batch_size, P^2•C, N] -> [batch_size, N, P^2•C]
print(f"Patch embedding sequence shape: {image_out_of_conv_flattened_reshaped.shape} -> [batch_size, num_patches, embedding_size]")
````

我们现在已经通过几个 PyTorch 层成功匹配了 ViT 架构中补丁嵌入层的输入和输出形状。

接下来，我们来看一下其中一个展平后的特征图的可视化效果吧？

````python
# 获取一个展平后的特征图
single_flattened_feature_map = image_out_of_conv_flattened_reshaped[:, :, 0] # 索引: (batch_size, number_of_patches, embedding_dimension)

# 以可视化方式绘制展平后的特征图
plt.figure(figsize=(22, 22))
plt.imshow(single_flattened_feature_map.detach().numpy())
plt.title(f"Flattened feature map shape: {single_flattened_feature_map.shape}")
plt.axis(False);
````

嗯，虽然展平后的特征图在视觉上看起来并不显眼，但这并不是我们关心的重点，它将成为补丁嵌入层的输出，并作为 ViT 架构其余部分的输入。

> **注意：** [原始的 Transformer 架构](https://arxiv.org/abs/1706.03762)是为文本处理设计的，而 Vision Transformer（ViT）架构的目标是将原始的 Transformer 应用于图像。这就是为什么 ViT 架构的输入是以这种方式处理的原因。我们实际上是将一张 2D 图像转换成一个 1D 的序列，类似于文本的处理方式。

接下来，我们来看一下展平后的特征图在张量中的表现形式吧？

````python
# 查看展平后的特征图作为张量的形式
single_flattened_feature_map, single_flattened_feature_map.requires_grad, single_flattened_feature_map.shape
````

我们已经将单张 2D 图像转化为一个 1D 可学习的嵌入向量（或者在 ViT 论文的图 1 中称为“展平补丁的线性投影”）。

### 4.5 将 ViT 补丁嵌入层转化为 PyTorch 模块

现在是时候将我们为创建补丁嵌入所做的所有工作封装成一个单独的 PyTorch 层了。

我们可以通过继承 `nn.Module` 来创建一个小的 PyTorch "模型"，以实现上面所有的步骤。

具体来说，我们将：
1. 创建一个名为 `PatchEmbedding` 的类，继承自 `nn.Module`（这样它就可以作为一个 PyTorch 层使用）。
2. 使用参数 `in_channels=3`，`patch_size=16`（适用于 ViT-Base）和 `embedding_dim=768`（这是 ViT-Base 中的 $D$，见表 1）来初始化类。
3. 使用 `nn.Conv2d()` 创建一个将图像转化为补丁的层（就像在 4.3 中所做的那样）。
4. 创建一个将补丁特征图展平为单一维度的层（就像在 4.4 中所做的那样）。
5. 定义一个 `forward()` 方法，接收输入并通过第 3 和第 4 步创建的层进行前向传播。
6. 确保输出形状符合 ViT 架构的要求输出形状（${N \times\left(P^{2} \cdot C\right)}$）。

让我们开始吧！

````python
# 1. 创建一个继承自 nn.Module 的类
class PatchEmbedding(nn.Module):
    """将 2D 输入图像转化为 1D 序列的可学习嵌入向量。

    参数:
        in_channels (int): 输入图像的颜色通道数。默认为 3。
        patch_size (int): 将输入图像转换为补丁的大小。默认为 16。
        embedding_dim (int): 将图像转化为的嵌入大小。默认为 768。
    """
    # 2. 使用合适的变量初始化类
    def __init__(self,
                 in_channels:int=3,
                 patch_size:int=16,
                 embedding_dim:int=768):
        super().__init__()

        # 3. 创建一个将图像转换为补丁的层
        self.patcher = nn.Conv2d(in_channels=in_channels,
                                 out_channels=embedding_dim,
                                 kernel_size=patch_size,
                                 stride=patch_size,
                                 padding=0)

        # 4. 创建一个将补丁特征图展平为单一维度的层
        self.flatten = nn.Flatten(start_dim=2, # 只展平特征图维度为一个单一向量
                                  end_dim=3)

    # 5. 定义前向传播方法
    def forward(self, x):
        # 创建断言来检查输入的形状是否正确
        image_resolution = x.shape[-1]
        assert image_resolution % patch_size == 0, f"输入图像大小必须能够被补丁大小整除，图像形状: {image_resolution}, 补丁大小: {patch_size}"

        # 执行前向传播
        x_patched = self.patcher(x)
        x_flattened = self.flatten(x_patched)
        # 6. 确保输出的形状是正确的顺序
        return x_flattened.permute(0, 2, 1) # 调整顺序，使嵌入维度位于最后一维 [batch_size, P^2•C, N] -> [batch_size, N, P^2•C]
````

`PatchEmbedding` 层创建完成！

现在让我们在一张图像上试一下。

````python
set_seeds()

# 创建 `PatchEmbedding` 层的实例
patchify = PatchEmbedding(in_channels=3,
                          patch_size=16,
                          embedding_dim=768)

# 通过单张图像进行测试
print(f"Input image shape: {image.unsqueeze(0).shape}")
patch_embedded_image = patchify(image.unsqueeze(0)) # 在第0维添加一个额外的批次维度，否则会出错
print(f"Output patch embedding shape: {patch_embedded_image.shape}")
````

输出形状与我们期望从补丁嵌入层中看到的理想输入和输出形状完全匹配：

* **输入：** 图像以 2D 形式开始，大小为 ${H \times W \times C}$。
* **输出：** 图像被转换为一个 1D 序列的展平 2D 补丁，大小为 ${N \times\left(P^{2} \cdot C\right)}$。

其中：
* $(H, W)$ 是原始图像的分辨率。
* $C$ 是通道数。
* $(P, P)$ 是每个图像补丁的分辨率（**补丁大小**）。
* $N=H W / P^{2}$ 是最终的补丁数，也就是变换器的有效输入序列长度。

我们已经复制了方程式 1 中的补丁嵌入，但还没有包括类标记（class token）和位置嵌入（position embedding）。

这些我们稍后会处理。

*我们的 `PatchEmbedding` 类（右）复制了 ViT 架构中的补丁嵌入（图 1 和方程 1，左）。然而，可学习的类嵌入和位置嵌入还没有创建，它们将在稍后添加。*

### 4.6 创建类标记嵌入

好了，我们已经完成了图像的补丁嵌入，现在开始处理类标记嵌入。

也就是方程 1 中的 $\mathbf{x}_\text {class }$。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-equation-1-the-class-token.png" width=900 alt="Vision Transformer 论文中图1和3.1节中的类标记嵌入突出显示"/>

*左图：来自 ViT 论文的图 1，突出显示了我们将重新创建的“分类标记”或 `[class]` 嵌入标记。右图：来自 ViT 论文的方程 1 和第 3.1 节，描述了与可学习类嵌入标记相关的内容。*

从 ViT 论文第 3.1 节的第二段，我们看到如下描述：

> 类似于 BERT 的 `[class]` 标记，我们将一个可学习的嵌入添加到嵌入补丁序列中 $\left(\mathbf{z}_{0}^{0}=\mathbf{x}_{\text {class }}\right)$，其在 Transformer 编码器输出时的状态 $\left(\mathbf{z}_{L}^{0}\right)$ 作为图像表示 $\mathbf{y}$ （方程 4）。

> **注：** [BERT](https://arxiv.org/abs/1810.04805)（双向编码器表示 Transformer）是最早使用 Transformer 架构在自然语言处理（NLP）任务中取得杰出成果的机器学习研究论文之一，也是第一个在序列开头添加 `[class]` 标记的论文，`class` 是指序列所属的“分类”类别。

所以，我们需要将“一个可学习的嵌入添加到嵌入补丁序列中”。

接下来，让我们查看我们在第 4.5 节创建的嵌入补丁序列张量及其形状。

````python
# 查看补丁嵌入和补丁嵌入形状
print(patch_embedded_image)
print(f"Patch embedding shape: {patch_embedded_image.shape} -> [batch_size, number_of_patches, embedding_dimension]")
````

为了将“可学习的嵌入添加到嵌入补丁序列前面”，我们需要创建一个形状为 `embedding_dimension`（$D$）的可学习嵌入，然后将其添加到 `number_of_patches` 维度中。

伪代码如下：

```python
patch_embedding = [image_patch_1, image_patch_2, image_patch_3...]
class_token = learnable_embedding
patch_embedding_with_class_token = torch.cat((class_token, patch_embedding), dim=1)
```

注意，连接操作 (`torch.cat()`) 是在 `dim=1`（即 `number_of_patches` 维度）上进行的。

接下来，我们来创建一个类标记的可学习嵌入。

为此，我们将获取批量大小和嵌入维度形状，然后创建一个形状为 `[batch_size, 1, embedding_dimension]` 的 `torch.ones()` 张量。

最后，我们将通过 `nn.Parameter()` 并设置 `requires_grad=True` 来使这个张量变为可学习的。

````python
# 获取批量大小和嵌入维度
batch_size = patch_embedded_image.shape[0]
embedding_dimension = patch_embedded_image.shape[-1]

# 创建类标记嵌入作为可学习的参数，其形状与嵌入维度 (D) 相同
class_token = nn.Parameter(torch.ones(batch_size, 1, embedding_dimension), # [batch_size, number_of_tokens, embedding_dimension]
                           requires_grad=True) # 确保该嵌入是可学习的

# 显示前 10 个 class_token 的例子
print(class_token[:, :, :10])

# 打印 class_token 的形状
print(f"Class token shape: {class_token.shape} -> [batch_size, number_of_tokens, embedding_dimension]")
````

> **注意：** 这里我们只是为了演示目的使用 `torch.ones()` 来创建类标记嵌入，实际上，你很可能会使用 `torch.randn()` 来创建类标记嵌入（因为机器学习就是要利用受控的随机性，通常你会从一个随机数开始，并随着时间的推移逐步改进它）。

可以看到，`class_token` 的 `number_of_tokens` 维度是 `1`，因为我们只想在补丁嵌入序列的开头插入一个类标记值。

现在我们已经得到了类标记嵌入，接下来将其添加到图像补丁序列 `patch_embedded_image` 中。

我们可以使用 [`torch.cat()`](https://pytorch.org/docs/stable/generated/torch.cat.html) 来实现，并设置 `dim=1`（这样 `class_token` 的 `number_of_tokens` 维度就会被添加到 `patch_embedded_image` 的 `number_of_patches` 维度的前面）。

````python
# 将类标记嵌入添加到补丁嵌入的前面
patch_embedded_image_with_class_embedding = torch.cat((class_token, patch_embedded_image),
                                                      dim=1)  # 在第一个维度上进行连接

# 打印带有类标记嵌入的补丁嵌入序列
print(patch_embedded_image_with_class_embedding)
print(f"Sequence of patch embeddings with class token prepended shape: {patch_embedded_image_with_class_embedding.shape} -> [batch_size, number_of_patches, embedding_dimension]")
````

很好！可学习的类标记已添加！

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-equation-1-prepending-the-learnable-class-token.png" width=900 alt="从一系列图像补丁嵌入开始，创建一个可学习的类标记，并将其添加到补丁嵌入序列的开头"/>

**回顾我们创建可学习类标记的过程**：我们首先通过对单张图像调用 `PatchEmbedding()` 创建了一系列图像补丁嵌入，然后创建了一个可学习的类标记，其值对应每个嵌入维度的一个值，接着将该标记添加到原始的补丁嵌入序列的开头。  
**注意：** 使用 `torch.ones()` 创建可学习类标记仅仅是为了演示目的，实际应用中通常会使用 `torch.randn()` 来创建它。

### 4.7 创建位置嵌入

我们已经有了类标记嵌入和补丁嵌入，那么如何创建位置嵌入呢？

也就是方程 1 中的 $\mathbf{E}_{\text {pos }}$，其中 $E$ 表示“嵌入”。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-equation-1-the-position-embeddings.png" width=900 alt="从视觉Transformer架构中提取位置嵌入，并与视觉Transformer论文的其他部分进行比较"/>

*左图：ViT 论文中的图 1，突出显示了我们即将重新创建的位置嵌入。右图：ViT 论文中的方程 1 和第 3.1 节，它们与位置嵌入相关。*

通过阅读 ViT 论文的第 3.1 节，我们来了解更多（加粗为重点）：

> 位置嵌入被添加到补丁嵌入中，以保留位置信息。我们使用**标准的可学习一维位置嵌入**，因为使用更高级的二维感知位置嵌入（附录 D.4）并未观察到显著的性能提升。生成的嵌入向量序列作为编码器的输入。

论文中提到的“保留位置信息”是指希望架构能够了解补丁的“顺序”。例如，补丁 2 紧随补丁 1 之后，补丁 3 紧随补丁 2 之后，以此类推。

在考虑图像中的内容时，位置信息非常重要（如果没有位置信息，展平的序列可能会被视为无序的，从而导致每个补丁之间没有关联）。

为了开始创建位置嵌入，让我们先查看当前的嵌入。

````python
# 查看添加了类标记嵌入的补丁嵌入序列
patch_embedded_image_with_class_embedding, patch_embedded_image_with_class_embedding.shape
````

方程 1 表明，位置嵌入 ($\mathbf{E}_{\text {pos }}$) 应具有形状 $(N + 1) \times D$：

$$\mathbf{E}_{\text {pos }} \in \mathbb{R}^{(N+1) \times D}$$

其中：
- $N = H W / P^{2}$ 是生成的补丁数量，同时也是 Transformer 的有效输入序列长度（即补丁数量）。
- $D$ 是**补丁嵌入**的维度，不同的 $D$ 值可以在表 1（嵌入维度）中找到。

幸运的是，这两个值我们已经有了。

接下来我们用 `torch.ones()` 创建一个可学习的一维嵌入来生成 $\mathbf{E}_{\text {pos }}$。

````python
# 计算 N（补丁数量）
number_of_patches = int((height * width) / patch_size**2)

# 获取嵌入维度
embedding_dimension = patch_embedded_image_with_class_embedding.shape[2]

# 创建可学习的一维位置嵌入
position_embedding = nn.Parameter(torch.ones(1,
                                             number_of_patches+1,
                                             embedding_dimension),
                                  requires_grad=True)  # 确保它是可学习的

# 显示前 10 个序列和前 10 个位置嵌入值，并检查位置嵌入的形状
print(position_embedding[:, :10, :10])
print(f"Position embeddding shape: {position_embedding.shape} -> [batch_size, number_of_patches, embedding_dimension]")
````

> **注意：** 这里仅用 `torch.ones()` 创建位置嵌入是为了演示目的，实际应用中通常会使用 `torch.randn()` 来创建位置嵌入（以随机数开始，并通过梯度下降进行优化）。

位置嵌入已创建！

现在将其添加到带有类标记的补丁嵌入序列中。

````python
# 将位置嵌入添加到补丁嵌入和类标记嵌入中
patch_and_position_embedding = patch_embedded_image_with_class_embedding + position_embedding
print(patch_and_position_embedding)
````

注意到嵌入张量中每个元素的值都增加了 1（这是因为位置嵌入是用 `torch.ones()` 创建的）。

> **注意：** 如果需要，我们可以将类标记嵌入和位置嵌入分别放入独立的层中。不过我们会在第 8 节中看到它们如何被整合到 ViT 架构的整体 `forward()` 方法中。

### 4.8 整合所有内容：从图像到嵌入

好了，我们已经在将输入图像转换为嵌入方面取得了很大进展，并且成功重现了 ViT 论文第 3.1 节中的方程 1：

$$
\begin{aligned}
\mathbf{z}_{0} &=\left[\mathbf{x}_{\text {class }} ; \mathbf{x}_{p}^{1} \mathbf{E} ; \mathbf{x}_{p}^{2} \mathbf{E} ; \cdots ; \mathbf{x}_{p}^{N} \mathbf{E}\right]+\mathbf{E}_{\text {pos }}, & & \mathbf{E} \in \mathbb{R}^{\left(P^{2} \cdot C\right) \times D}, \mathbf{E}_{\text {pos }} \in \mathbb{R}^{(N+1) \times D}
\end{aligned}
$$

现在，让我们将所有内容整合到一个代码单元中，完成从输入图像 ($\mathbf{x}$) 到输出嵌入 ($\mathbf{z}_0$) 的过程。

实现步骤如下：
1. 设置补丁大小（我们使用 `16`，因为它在论文和 ViT-Base 中被广泛采用）。
2. 获取一张单一图像，打印其形状并记录其高度和宽度。
3. 为单一图像添加一个批次维度，使其与 `PatchEmbedding` 层兼容。
4. 创建一个 `PatchEmbedding` 层（第 4.5 节中创建的版本），其中 `patch_size=16` 和 `embedding_dim=768`（根据 ViT-Base 的表 1）。
5. 将单一图像传入第 4 步中的 `PatchEmbedding` 层，生成补丁嵌入序列。
6. 按照第 4.6 节的方式创建一个类标记嵌入。
7. 将类标记嵌入添加到第 5 步生成的补丁嵌入前面。
8. 按照第 4.7 节的方式创建位置嵌入。
9. 将位置嵌入添加到第 7 步生成的类标记和补丁嵌入中。

我们还将通过 `set_seeds()` 设置随机种子，并在过程中打印不同张量的形状以便检查。

````python
set_seeds()

# 1. 设置补丁大小
patch_size = 16

# 2. 打印原始图像张量的形状并获取图像尺寸
print(f"Image tensor shape: {image.shape}")
height, width = image.shape[1], image.shape[2]

# 3. 获取图像张量并添加批次维度
x = image.unsqueeze(0)
print(f"Input image with batch dimension shape: {x.shape}")

# 4. 创建补丁嵌入层
patch_embedding_layer = PatchEmbedding(in_channels=3,
                                       patch_size=patch_size,
                                       embedding_dim=768)

# 5. 将图像传入补丁嵌入层
patch_embedding = patch_embedding_layer(x)
print(f"Patching embedding shape: {patch_embedding.shape}")

# 6. 创建类标记嵌入
batch_size = patch_embedding.shape[0]
embedding_dimension = patch_embedding.shape[-1]
class_token = nn.Parameter(torch.ones(batch_size, 1, embedding_dimension),
                           requires_grad=True) # make sure it's learnable
print(f"Class token embedding shape: {class_token.shape}")

# 7. 将类标记嵌入添加到补丁嵌入前面
patch_embedding_class_token = torch.cat((class_token, patch_embedding), dim=1)
print(f"Patch embedding with class token shape: {patch_embedding_class_token.shape}")

# 8. 创建位置嵌入
number_of_patches = int((height * width) / patch_size**2)
position_embedding = nn.Parameter(torch.ones(1, number_of_patches+1, embedding_dimension),
                                  requires_grad=True) # 确保它是可学习的

# 9. 将位置嵌入添加到带有类标记的补丁嵌入中
patch_and_position_embedding = patch_embedding_class_token + position_embedding
print(f"Patch and position embedding shape: {patch_and_position_embedding.shape}")
````

## 5. 方程2：多头注意力（MSA）

我们已经将输入数据分割成补丁并完成了嵌入，现在进入ViT架构的下一部分。

首先，我们将Transformer Encoder部分拆分为两个部分（从简单开始，根据需要逐步扩展）。

第一个部分是方程2，第二个部分是方程3。

回顾方程2的内容：

$$
\begin{aligned}
\mathbf{z}_{\ell}^{\prime} &=\operatorname{MSA}\left(\operatorname{LN}\left(\mathbf{z}_{\ell-1}\right)\right)+\mathbf{z}_{\ell-1}, & & \ell=1 \ldots L
\end{aligned}
$$

这表示一个多头注意力（MSA）层嵌套在一个LayerNorm（LN）层中，并带有一个残差连接（将输入添加到该层的输出中）。

我们将方程2称为“MSA模块”。

<img src="https://raw.githubusercontent.com/mrdbourke/pytorch-deep-learning/main/images/08-vit-paper-equation-2-msa-block-mapped-to-equation.png" alt="从ViT论文的方程2映射到图1的ViT架构图的示例" width=900/>

在许多研究论文中提到的层已经被现代深度学习框架（如PyTorch）实现。

具体来说，为了用PyTorch代码复现这些层和残差连接，可以使用以下方法：
- **多头自注意力（MSA）** - 使用 [`torch.nn.MultiheadAttention()`](https://pytorch.org/docs/stable/generated/torch.nn.MultiheadAttention.html)。
- **归一化（LN或LayerNorm）** - 使用 [`torch.nn.LayerNorm()`](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html)。
- **残差连接** - 将输入加到输出上（我们将在第7.1节创建完整的Transformer Encoder模块时看到这一点）。

### 5.1 LayerNorm（LN）层

[层归一化（Layer Normalization）](https://paperswithcode.com/method/layer-normalization)（`torch.nn.LayerNorm()` 或简称为 Norm、LayerNorm 或 LN）是对输入数据的最后一个维度进行归一化处理。

关于 `torch.nn.LayerNorm()` 的正式定义，可以参考 [PyTorch 文档](https://pytorch.org/docs/stable/generated/torch.nn.LayerNorm.html)。

PyTorch 的 `torch.nn.LayerNorm()` 的主要参数是 `normalized_shape`，我们可以将其设置为需要归一化的维度大小（在我们的 ViT-Base 中为 $D$ 或 `768`）。

**LayerNorm 的作用是什么？**

层归一化可以帮助提高训练速度，并增强模型的泛化能力（适应未见过数据的能力）。

我更倾向于将各种归一化操作理解为“让数据变得格式相似”或者“让数据样本的分布变得一致”。

假设你正试图爬一段每一级高度和长度都不相同的楼梯。

你每迈一步都需要调整对吧？

而且你在某一级楼梯上的调整经验不一定适用于下一步，因为每一级楼梯都不同，这就增加了你完成这段楼梯的时间。

归一化（包括层归一化）就像是让所有楼梯的高度和长度都相同，而这些楼梯对应的就是你的数据样本。

就像你在高度和长度相似的楼梯上行走比在不规则的楼梯上容易得多一样，神经网络在处理分布（均值和标准差）相似的数据样本时，也会比处理分布各异的数据样本更高效。

### 5.2 多头自注意力（MSA）层

自注意力和多头注意力（多次应用自注意力机制）的强大之处首次在[*Attention is All You Need*](https://arxiv.org/abs/1706.03762)论文中展示，该论文提出了最初的Transformer架构。

最初设计用于文本输入，自注意力机制会对一系列单词进行计算，确定哪个单词应该对另一个单词给予更多“关注”。

换句话说，在句子“The dog jumped over the fence”中，也许“dog”与“jumped”和“fence”关系更紧密。

虽然这是一个简化的例子，但对于图像来说，原理是相同的。

由于我们的输入是图像补丁的序列而不是单词，自注意力机制和多头注意力将计算图像的哪个补丁与另一个补丁最相关，最终形成对图像的学习表示。

最重要的是，该层可以根据数据自主完成这些计算（我们并不需要明确告诉它要学习什么样的模式）。

如果使用MSA机制形成的学习表示效果良好，我们将能从模型的性能中看到结果。

关于Transformer架构和注意力机制，有许多在线资源可供学习，例如Jay Alammar的优秀文章：[Illustrated Transformer](https://jalammar.github.io/illustrated-transformer/)和[Illustrated Attention](https://jalammar.github.io/visualizing-neural-machine-translation-mechanics-of-seq2seq-models-with-attention/)。

我们将重点放在使用PyTorch实现现有的MSA实现，而不是从头创建自己的版本。

然而，在ViT论文的附录A中，可以找到ViT中MSA实现的正式定义：

上图突出显示了输入MSA层的三重嵌入输入。

这被称为**查询（query）、键（key）、值（value）输入**，简称为**qkv**，这是自注意力机制的核心。

在我们的例子中，三重嵌入输入将是归一化层输出的三种版本，分别对应于查询、键和值。

也就是我们在第4.8节中创建的归一化图像补丁和位置嵌入的三种版本。

我们可以通过PyTorch的`torch.nn.MultiheadAttention()`实现MSA层，其参数包括：
- **`embed_dim`**：来自表1的嵌入维度（隐藏大小 $D$）。
- **`num_heads`**：使用的注意力头数（“多头”的由来），该值也在表1中（Heads）。
- **`dropout`**：是否对注意力层应用dropout（根据附录B.1，qkv投影后未使用dropout）。
- **`batch_first`**：批次维度是否在最前面（是的，在这里是批次优先）。

### 5.3 用 PyTorch 层复现方程 2

让我们将关于 LayerNorm（LN）和多头自注意力（MSA）层在方程 2 中的讨论付诸实践。

为此，我们将：

1. 创建一个名为 `MultiheadSelfAttentionBlock` 的类，继承自 `torch.nn.Module`。
2. 使用 ViT 论文表 1 中 ViT-Base 模型的超参数来初始化该类。
3. 使用 `torch.nn.LayerNorm()` 创建一个层归一化（LN）层，其 `normalized_shape` 参数设置为我们的嵌入维度（表 1 中的 $D$）。
4. 创建一个多头自注意力（MSA）层，设置适当的 `embed_dim`、`num_heads`、`dropout` 和 `batch_first` 参数。
5. 为我们的类创建一个 `forward()` 方法，将输入依次通过 LN 层和 MSA 层。

````python
# 1. 创建一个继承自 nn.Module 的类
class MultiheadSelfAttentionBlock(nn.Module):
    """
    创建一个多头自注意力模块（简称 "MSA block"）。
    """
    # 2. 使用表 1 中的超参数初始化类
    def __init__(self,
                 embedding_dim: int = 768,  # ViT-Base 模型的隐藏维度 D（表 1 中的值）
                 num_heads: int = 12,       # ViT-Base 模型的注意力头数（表 1 中的值）
                 attn_dropout: float = 0):  # 论文中似乎没有在 MSABlocks 中使用 dropout
        super().__init__()

        # 3. 创建 Norm 层（LN）
        self.layer_norm = nn.LayerNorm(normalized_shape=embedding_dim)

        # 4. 创建多头注意力（MSA）层
        self.multihead_attn = nn.MultiheadAttention(embed_dim=embedding_dim,
                                                    num_heads=num_heads,
                                                    dropout=attn_dropout,
                                                    batch_first=True)  # 批次维度是否在最前？

    # 5. 创建 forward() 方法将数据传递到各层中
    def forward(self, x):
        x = self.layer_norm(x)  # 通过归一化层
        attn_output, _ = self.multihead_attn(query=x,  # 查询嵌入
                                             key=x,    # 键嵌入
                                             value=x,  # 值嵌入
                                             need_weights=False)  # 是否需要注意力权重，还是只需要层输出？
        return attn_output
````

> **注意：** 与图 1 不同，我们的 `MultiheadSelfAttentionBlock` 尚未包含跳跃或残差连接（即方程 2 中的 "$+\mathbf{z}_{\ell-1}$"）。我们将在第 7.1 节创建完整的 Transformer Encoder 时加入这一部分。

MSA 模块已创建！

现在尝试一下：创建一个 `MultiheadSelfAttentionBlock` 的实例，并将第 4.8 节中生成的 `patch_and_position_embedding` 变量传递进去。

````python
# 创建一个 MSA 模块的实例
multihead_self_attention_block = MultiheadSelfAttentionBlock(embedding_dim=768,  # 来自表 1 的嵌入维度
                                                             num_heads=12)       # 来自表 1 的注意力头数

# 将图像补丁和位置嵌入传递到 MSA 模块中
patched_image_through_msa_block = multihead_self_attention_block(patch_and_position_embedding)
print(f"Input shape of MSA block: {patch_and_position_embedding.shape}")
print(f"Output shape MSA block: {patched_image_through_msa_block.shape}")
````

### 6.1 MLP 层

[MLP](https://en.wikipedia.org/wiki/Multilayer_perceptron)（多层感知机）是一个较为宽泛的术语，可以指几乎任何*多个*层的组合（这也是“多层感知机”中“多层”的含义）。

通常，它遵循以下模式：

`线性层 -> 非线性层 -> 线性层 -> 非线性层`

在 ViT 论文中，MLP 的结构在第 3.1 节中有明确定义：

> MLP 包含两个带有 GELU 非线性激活的层。

其中，“两个层”指的是线性层（在 PyTorch 中是 [`torch.nn.Linear()`](https://pytorch.org/docs/stable/generated/torch.nn.Linear.html)），“GELU 非线性激活”是指 GELU（Gaussian Error Linear Units，高斯误差线性单元）非线性激活函数（在 PyTorch 中是 [`torch.nn.GELU()`](https://pytorch.org/docs/stable/generated/torch.nn.GELU.html)）。

> **注意：** 线性层（`torch.nn.Linear()`）有时也被称为“密集层”（dense layer）或“前馈层”（feedforward layer）。有些论文甚至会同时使用这三个术语来描述同一概念（如在 ViT 论文中）。

关于 MLP 模块的另一个隐藏细节在附录 B.1（训练）中才出现：

> 表 3 总结了我们不同模型的训练设置。...Dropout（如果使用）会在**每个密集层后应用，除了 qkv 投影层以及在位置嵌入和补丁嵌入相加之后直接应用的部分。**

这意味着在 MLP 模块中，每个线性层（或密集层）后都需要一个 dropout 层（在 PyTorch 中是 [`torch.nn.Dropout()`](https://pytorch.org/docs/stable/generated/torch.nn.Dropout.html)）。

对于 ViT-Base 模型，dropout 的值可以在论文的表 3 中找到（`dropout=0.1`）。

基于这些信息，我们的 MLP 模块结构如下：

`层归一化 -> 线性层 -> 非线性层 -> Dropout -> 线性层 -> Dropout`

其中，线性层的超参数可从表 1 中获取（MLP 大小表示线性层之间的隐藏单元数，隐藏大小 $D$ 表示 MLP 模块的输出大小）。

### 6.2 使用 PyTorch 层复现方程 3

让我们将关于 LayerNorm（LN）和 MLP 层在方程 3 中的讨论付诸实践。

为此，我们将：

1. 创建一个名为 `MLPBlock` 的类，继承自 `torch.nn.Module`。
2. 使用 ViT 论文表 1 和表 3 中的超参数来初始化该类（针对 ViT-Base 模型）。
3. 使用 `torch.nn.LayerNorm()` 创建一个层归一化（LN）层，其 `normalized_shape` 参数设置为嵌入维度（表 1 中的 $D$）。
4. 使用 `torch.nn.Linear()`、`torch.nn.Dropout()` 和 `torch.nn.GELU()` 创建一系列 MLP 层，超参数取自表 1 和表 3。
5. 为类创建一个 `forward()` 方法，将输入依次传递通过 LN 层和 MLP 层。

````python
# 1. 创建一个继承自 nn.Module 的类
class MLPBlock(nn.Module):
    """创建一个带有层归一化的多层感知机模块（简称 "MLP block"）。"""
    # 2. 使用表 1 和表 3 中的超参数初始化类
    def __init__(self,
                 embedding_dim: int = 768,  # 来自表 1 的隐藏维度 D（针对 ViT-Base）
                 mlp_size: int = 3072,      # 来自表 1 的 MLP 大小（针对 ViT-Base）
                 dropout: float = 0.1):     # 来自表 3 的 Dropout 值（针对 ViT-Base）
        super().__init__()

        # 3. 创建 Norm 层（LN）
        self.layer_norm = nn.LayerNorm(normalized_shape=embedding_dim)

        # 4. 创建多层感知机（MLP）层
        self.mlp = nn.Sequential(
            nn.Linear(in_features=embedding_dim,  # 输入维度为嵌入维度
                      out_features=mlp_size),    # 输出维度为 MLP 大小
            nn.GELU(),                           # "MLP 包含两个带有 GELU 非线性激活的层"（第 3.1 节）
            nn.Dropout(p=dropout),               # Dropout 在每个密集层之后应用
            nn.Linear(in_features=mlp_size,      # 输入维度需要与上一层的输出维度相同
                      out_features=embedding_dim), # 输出维度回到嵌入维度
            nn.Dropout(p=dropout)                # 再次应用 Dropout
        )

    # 5. 创建 forward() 方法，将数据传递到各层中
    def forward(self, x):
        x = self.layer_norm(x)  # 通过归一化层
        x = self.mlp(x)         # 通过 MLP 层
        return x
````

> **注意：** 与图 1 不同，我们的 `MLPBlock()` 尚未包含跳跃或残差连接（即方程 3 中的 "$+\mathbf{z}_{\ell}^{\prime}$"）。我们将在后续创建完整的 Transformer 编码器时加入这一部分。

MLPBlock 类已创建！

现在尝试一下：创建一个 `MLPBlock` 的实例，并将第 5.3 节中生成的 `patched_image_through_msa_block` 变量传递进去。

````python
# 创建一个 MLPBlock 的实例
mlp_block = MLPBlock(embedding_dim=768,  # 来自表 1 的嵌入维度
                     mlp_size=3072,     # 来自表 1 的 MLP 大小
                     dropout=0.1)       # 来自表 3 的 Dropout 值

# 将 MSABlock 的输出传递到 MLPBlock 中
patched_image_through_mlp_block = mlp_block(patched_image_through_msa_block)
print(f"Input shape of MLP block: {patched_image_through_msa_block.shape}")
print(f"Output shape MLP block: {patched_image_through_mlp_block.shape}")
````

注意，当数据通过 MLP 模块时，输入和输出的形状再次保持不变。

然而，当数据通过 MLP 模块中的 `nn.Linear()` 层时，形状会发生变化（首先扩展到表 1 中的 MLP 大小，然后压缩回表 1 中的隐藏大小 $D$）。

## 7. 创建 Transformer 编码器

现在是时候将 `MultiheadSelfAttentionBlock`（方程 2）和 `MLPBlock`（方程 3）组合起来，创建 ViT 架构的 Transformer 编码器了。

在深度学习中，[“编码器”或“自动编码器”](https://paperswithcode.com/method/autoencoder) 通常指的是一组可以“编码”输入（将其转化为某种形式的数值表示）的层。

在我们的例子中，Transformer 编码器将按照 ViT 论文第 3.1 节的描述，通过交替的 MSA 模块和 MLP 模块层，将补丁图像嵌入编码为一个学习表示：

> Transformer 编码器（Vaswani 等，2017）由交替的多头自注意力（MSA，见附录 A）和 MLP 模块（方程 2、3）组成。**每个模块之前应用层归一化（LN）**，**每个模块之后添加残差连接**（Wang 等，2019；Baevski 和 Auli，2019）。

我们已经创建了 MSA 和 MLP 模块，但残差连接怎么办？

[残差连接](https://paperswithcode.com/method/residual-connection)（也称为跳跃连接）最初在 [*Deep Residual Learning for Image Recognition*](https://arxiv.org/abs/1512.03385v1) 论文中被提出，其实现方式是将一层的输入与其后续输出相加。

在某些情况下，这个后续输出可能是来自一个或多个后续层。

在 ViT 架构中，残差连接的实现是，将 MSA 模块的输入加回到其输出中，然后再传递给 MLP 模块。

同样的操作也适用于 MLP 模块，在其输出传递给下一个 Transformer 编码器模块之前。

伪代码如下：

`x_input -> MSA_block -> [MSA_block_output + x_input] -> MLP_block -> [MLP_block_output + MSA_block_output + x_input] -> ...`

**这有什么作用？**

残差连接的一个主要作用是防止权重值和梯度更新变得过小，从而使得更深的网络能够更好地训练，并允许学习更深层次的表示。

> **注意：** 著名的计算机视觉架构“ResNet”之所以得名，正是因为引入了*残*差连接。你可以在 [`torchvision.models`](https://pytorch.org/vision/stable/models.html) 中找到许多预训练的 ResNet 架构版本。

### 7.1 使用自定义层创建 Transformer 编码器

说了这么多，现在来实际操作一下，通过组合之前创建的层，用 PyTorch 构建一个 ViT Transformer 编码器。

为此，我们将：

1. 创建一个名为 `TransformerEncoderBlock` 的类，继承自 `torch.nn.Module`。
2. 使用 ViT 论文表 1 和表 3 中的超参数（针对 ViT-Base 模型）来初始化类。
3. 使用第 5.2 节中创建的 `MultiheadSelfAttentionBlock` 实例化一个 MSA 模块（用于方程 2），并设置适当的参数。
4. 使用第 6.2 节中创建的 `MLPBlock` 实例化一个 MLP 模块（用于方程 3），并设置适当的参数。
5. 为 `TransformerEncoderBlock` 类创建一个 `forward()` 方法。
6. 为 MSA 模块创建一个残差连接（对应方程 2）。
7. 为 MLP 模块创建一个残差连接（对应方程 3）。

````python
# 1. 创建一个继承自 nn.Module 的类
class TransformerEncoderBlock(nn.Module):
    """创建一个 Transformer 编码器模块。"""
    # 2. 使用表 1 和表 3 中的超参数初始化类
    def __init__(self,
                 embedding_dim: int = 768,  # 来自表 1 的隐藏维度 D（针对 ViT-Base）
                 num_heads: int = 12,       # 来自表 1 的注意力头数（针对 ViT-Base）
                 mlp_size: int = 3072,      # 来自表 1 的 MLP 大小（针对 ViT-Base）
                 mlp_dropout: float = 0.1,  # 表 3 中密集层的 Dropout 值（针对 ViT-Base）
                 attn_dropout: float = 0):  # 注意力层的 Dropout 值
        super().__init__()

        # 3. 创建 MSA 模块（对应方程 2）
        self.msa_block = MultiheadSelfAttentionBlock(embedding_dim=embedding_dim,
                                                     num_heads=num_heads,
                                                     attn_dropout=attn_dropout)

        # 4. 创建 MLP 模块（对应方程 3）
        self.mlp_block = MLPBlock(embedding_dim=embedding_dim,
                                  mlp_size=mlp_size,
                                  dropout=mlp_dropout)

    # 5. 创建 forward() 方法
    def forward(self, x):

        # 6. 为 MSA 模块创建残差连接（将输入加到输出上）
        x = self.msa_block(x) + x

        # 7. 为 MLP 模块创建残差连接（将输入加到输出上）
        x = self.mlp_block(x) + x

        return x
````

$$

别担心，我们可以将方程 4 整合到整体的 ViT 架构类中。

我们只需要一个 `torch.nn.LayerNorm()` 层和一个 `torch.nn.Linear()` 层，将 Transformer 编码器输出中第 0 个索引的 logits ($\mathbf{z}_{L}^{0}$) 转换为目标类别数即可。

为了创建完整架构，我们还需要将多个 `TransformerEncoderBlock` 堆叠在一起，可以通过将这些模块的列表传递给 `torch.nn.Sequential()` 来实现（这会创建一组顺序的 `TransformerEncoderBlock`）。

创建 ViT 将是我们目前最大的代码块，但我们能做到！

### 实现 ViT 的步骤：

1. 创建一个名为 `ViT` 的类，继承自 `torch.nn.Module`。
2. 使用 ViT 论文表 1 和表 3 的超参数（针对 ViT-Base 模型）初始化类。
3. 确保图像大小可以被补丁大小整除（图像应能均匀划分为补丁）。
4. 使用公式 $N=H W / P^{2}$ 计算补丁数量，其中 $H$ 是图像高度，$W$ 是图像宽度，$P$ 是补丁大小。
5. 创建一个可学习的类别嵌入 token（方程 1），如第 4.6 节所述。
6. 创建一个可学习的位置嵌入向量（方程 1），如第 4.7 节所述。
7. 按照 ViT 论文附录 B.1 的讨论，设置嵌入 dropout 层。
8. 使用第 4.5 节中的 `PatchEmbedding` 类创建补丁嵌入层。
9. 使用第 7.1 节创建的 `TransformerEncoderBlock` 列表，通过 `torch.nn.Sequential()` 创建一系列 Transformer 编码器模块（方程 2 和 3）。
10. 创建 MLP Head（也称为分类器 head，或方程 4），使用 `torch.nn.LayerNorm()` 和 `torch.nn.Linear(out_features=num_classes)`（其中 `num_classes` 是目标类别数）组成的层传递给 `torch.nn.Sequential()`。
11. 创建一个 `forward()` 方法，用于接收输入。
12. 获取输入的批量大小（形状的第一个维度）。
13. 使用第 8 步中创建的补丁嵌入层生成补丁嵌入（方程 1）。
14. 使用第 5 步中创建的类别 token 嵌入，并通过 [`torch.Tensor.expand()`](https://pytorch.org/docs/stable/generated/torch.Tensor.expand.html) 将其扩展到第 11 步中找到的批量大小（方程 1）。
15. 使用 [`torch.cat()`](https://pytorch.org/docs/stable/generated/torch.cat.html) 将第 14 步创建的类别 token 嵌入与第 12 步生成的补丁嵌入的第一个维度连接（方程 1）。
16. 将第 14 步创建的位置嵌入添加到第 15 步生成的补丁和类别 token 嵌入中（方程 1）。
17. 将第 16 步生成的补丁和位置嵌入通过第 7 步创建的 dropout 层。
18. 将第 17 步生成的补丁和位置嵌入通过第 9 步创建的 Transformer 编码器模块堆栈（方程 2 和 3）。
19. 将第 17 步 Transformer 编码器堆栈输出的第 0 索引通过第 10 步创建的分类器 Head（方程 4）。
20. 大声欢呼并跳舞！我们刚刚构建了一个 Vision Transformer！

准备好了吗？

开始吧！

````python
# 1. 创建一个继承自 nn.Module 的 ViT 类
class ViT(nn.Module):
    """默认使用 ViT-Base 超参数创建 Vision Transformer 架构。"""
    # 2. 使用表 1 和表 3 的超参数初始化类
    def __init__(self,
                 img_size: int = 224, # ViT 论文表 3 中的训练分辨率
                 in_channels: int = 3, # 输入图像的通道数
                 patch_size: int = 16, # 补丁大小
                 num_transformer_layers: int = 12, # ViT-Base 表 1 中的层数
                 embedding_dim: int = 768, # ViT-Base 表 1 中的隐藏维度 D
                 mlp_size: int = 3072, # ViT-Base 表 1 中的 MLP 大小
                 num_heads: int = 12, # ViT-Base 表 1 中的注意力头数
                 attn_dropout: float = 0, # 注意力投影的 Dropout
                 mlp_dropout: float = 0.1, # MLP 层的 Dropout
                 embedding_dropout: float = 0.1, # 补丁和位置嵌入的 Dropout
                 num_classes: int = 1000): # 默认用于 ImageNet，可自定义
        super().__init__() # 别忘了调用 super().__init__()

        # 3. 确保图像大小可以被补丁大小整除
        assert img_size % patch_size == 0, f"Image size must be divisible by patch size, image size: {img_size}, patch size: {patch_size}."

        # 4. 计算补丁数量（高度 * 宽度 / 补丁面积）
        self.num_patches = (img_size * img_size) // patch_size**2

        # 5. 创建可学习的类别嵌入（需放在补丁嵌入序列前）
        self.class_embedding = nn.Parameter(data=torch.randn(1, 1, embedding_dim),
                                            requires_grad=True)

        # 6. 创建可学习的位置嵌入
        self.position_embedding = nn.Parameter(data=torch.randn(1, self.num_patches + 1, embedding_dim),
                                               requires_grad=True)

        # 7. 创建嵌入的 Dropout 值
        self.embedding_dropout = nn.Dropout(p=embedding_dropout)

        # 8. 创建补丁嵌入层
        self.patch_embedding = PatchEmbedding(in_channels=in_channels,
                                              patch_size=patch_size,
                                              embedding_dim=embedding_dim)

        # 9. 创建 Transformer 编码器模块（使用 nn.Sequential() 堆叠 Transformer 编码器模块）
        # 注意："*" 表示“全部”
        self.transformer_encoder = nn.Sequential(*[
            TransformerEncoderBlock(embedding_dim=embedding_dim,
                                    num_heads=num_heads,
                                    mlp_size=mlp_size,
                                    mlp_dropout=mlp_dropout)
            for _ in range(num_transformer_layers)
        ])

        # 10. 创建分类器 Head
        self.classifier = nn.Sequential(
            nn.LayerNorm(normalized_shape=embedding_dim),
            nn.Linear(in_features=embedding_dim,
                      out_features=num_classes)
        )

    # 11. 创建 forward() 方法
    def forward(self, x):

        # 12. 获取批量大小
        batch_size = x.shape[0]

        # 13. 创建类别 token 嵌入并扩展以匹配批量大小（方程 1）
        class_token = self.class_embedding.expand(batch_size, -1, -1) # "-1" 表示推断该维度

        # 14. 创建补丁嵌入（方程 1）
        x = self.patch_embedding(x)

        # 15. 将类别嵌入与补丁嵌入拼接（方程 1）
        x = torch.cat((class_token, x), dim=1)

        # 16. 将位置嵌入添加到补丁嵌入中（方程 1）
        x = self.position_embedding + x

        # 17. 运行嵌入 Dropout（附录 B.1）
        x = self.embedding_dropout(x)

        # 18. 将补丁、位置和类别嵌入通过 Transformer 编码器层（方程 2 和 3）
        x = self.transformer_encoder(x)

        # 19. 将第 0 索引的 logit 通过分类器（方程 4）
        x = self.classifier(x[:, 0]) # 对批量中的每个样本取第 0 索引

        return x
````

> **Training & Fine-tuning.** 我们使用 Adam（Kingma & Ba, 2015），设置 $\beta_{1}=0.9, \beta_{2}=0.999$，批量大小为 4096，并应用高权重衰减 $0.1$ 训练所有模型，包括 ResNets。我们发现这对所有模型的迁移学习都非常有用（附录 D.1 显示，与常见做法不同，Adam 在我们的设置中比 SGD 在 ResNets 上略有优势）。

由此可以看出，作者选择使用 “Adam” 优化器（[`torch.optim.Adam()`](https://pytorch.org/docs/stable/generated/torch.optim.Adam.html#torch.optim.Adam)），而不是 SGD（随机梯度下降，[`torch.optim.SGD()`](https://pytorch.org/docs/stable/generated/torch.optim.SGD.html#torch.optim.SGD)）。

论文中将 Adam 的 $\beta$ 值设置为 $\beta_{1}=0.9, \beta_{2}=0.999$，这是 `torch.optim.Adam(betas=(0.9, 0.999))` 的默认值。

作者还提到使用 [权重衰减](https://paperswithcode.com/method/weight-decay)（在优化过程中逐渐减小权重值以防止过拟合），我们可以通过 `torch.optim.Adam(weight_decay=0.3)` 设置该参数（根据 ViT-* 在 ImageNet-1k 上训练的设置）。

根据表 3 的描述（ViT-* 在 ImageNet-1k 上训练的设置），我们将优化器的学习率设置为 0.003。

如前所述，由于硬件限制，我们将使用比 4096 更小的批量大小（如果你有更大的 GPU，可以随意增加批量大小）。

### 9.2 创建损失函数

奇怪的是，在 ViT 论文中搜索“loss”（损失）、“loss function”（损失函数）或“criterion”（准则）均没有结果。

然而，由于我们处理的目标问题是多分类问题（与 ViT 论文一致），我们将使用 [`torch.nn.CrossEntropyLoss()`](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html)。

### 9.3 训练我们的 ViT 模型

好了，现在我们已经知道要使用的优化器和损失函数，让我们为训练 ViT 设置训练代码。

````python
from going_modular.going_modular import engine

# 使用 ViT 论文中的超参数设置优化器以优化 ViT 模型的参数
optimizer = torch.optim.Adam(params=vit.parameters(),
                             lr=3e-3, # 表 3 中 ViT-* ImageNet-1k 的基础学习率
                             betas=(0.9, 0.999), # 默认值，同时也在 ViT 论文第 4.1 节（训练与微调）中提到
                             weight_decay=0.3) # 来自 ViT 论文第 4.1 节和表 3 中 ViT-* ImageNet-1k 的权重衰减

# 设置多分类问题的损失函数
loss_fn = torch.nn.CrossEntropyLoss()

# 设置随机种子
set_seeds()

# 训练模型，并将训练结果保存到字典中
results = engine.train(model=vit,
                       train_dataloader=train_dataloader,
                       test_dataloader=test_dataloader,
                       optimizer=optimizer,
                       loss_fn=loss_fn,
                       epochs=10,
                       device=device)
````

### 9.4 我们的训练设置中缺少什么

原始的 ViT 架构在多个图像分类基准测试上取得了出色的结果（发布时与许多最新方法表现相当甚至更好）。

原因可能有很多，但主要问题在于**规模**。

原始 ViT 论文使用的数据量远大于我们所用的数据量（在深度学习中，更多的数据通常是好事），并且训练计划也更长（参见表 3）。

| Epochs（训练轮数） | 7（最大数据集），90，300（ImageNet） | 10 |

| [学习率预热](https://paperswithcode.com/method/linear-warmup) | 10k 步（表 3） | 无 |

### 10.1 为什么要使用预训练模型？

许多现代机器学习研究论文的一个重要特点是，大部分结果是通过大规模数据集和强大的计算资源获得的。

在当今的机器学习领域，原始完全训练的 ViT 模型可能不会被视为“超大规模”训练设置（模型规模正变得越来越大）。

阅读 ViT 论文第 4.2 节：

> 最后，在公开的 ImageNet-21k 数据集上预训练的 ViT-L/16 模型在大多数数据集上表现良好，同时所需的预训练资源更少：它可以在大约 **30 天** 内使用标准的 8 核云 TPUv3 进行训练。

截至 2022 年 7 月，[Google Cloud 上租用 TPUv3](https://cloud.google.com/tpu/pricing)（第 3 版张量处理单元）的价格为每小时 8 美元。

连续 30 天的租用费用为 **$5,760 美元**。

这样的费用（包括金钱和时间）可能对一些较大的研究团队或企业来说是可行的，但对许多人来说却不是。

因此，通过诸如 [`torchvision.models`](https://pytorch.org/vision/stable/models.html)、[`timm`（Torch Image Models）库](https://github.com/rwightman/pytorch-image-models)、[HuggingFace Hub](https://huggingface.co/models) 或直接从论文作者处获得预训练模型资源是非常有意义的（越来越多的机器学习研究人员开始发布他们研究论文的代码和预训练模型，我非常支持这种趋势，许多这样的资源可以在 [Paperswithcode.com](https://paperswithcode.com/) 上找到）。

如果你的重点是利用特定模型架构的优势，而不是创建自定义架构，我强烈推荐使用预训练模型。

具体来说，我们将：

1. 获取在 ImageNet-1k 上训练的 ViT-Base 的预训练权重，使用 [`torchvision.models.ViT_B_16_Weights.DEFAULT`](https://pytorch.org/vision/stable/models/generated/torchvision.models.vit_b_16.html#torchvision.models.ViT_B_16_Weights)（`DEFAULT` 表示最佳可用权重）。

3. 冻结第 2 步创建的 ViT 模型基础部分的所有参数，通过将其 `requires_grad` 属性设置为 `False`。
4. 更新第 2 步创建的 ViT 模型的分类器 head，将 `out_features` 更改为我们的类别数量（披萨、牛排、寿司）。

我们在 06. PyTorch Transfer Learning 的 [第 3.2 节：设置预训练模型](https://www.learnpytorch.io/06_pytorch_transfer_learning/#32-setting-up-a-pretrained-model) 和 [第 3.4 节：冻结基础模型并更改输出层以满足需求](https://www.learnpytorch.io/06_pytorch_transfer_learning/#34-freezing-the-base-model-and-changing-the-output-layer-to-suit-our-needs) 中已经介绍过类似的步骤。

````python
# 1. 获取 ViT-Base 的预训练权重
pretrained_vit_weights = torchvision.models.ViT_B_16_Weights.DEFAULT # 需要 torchvision >= 0.13，"DEFAULT" 表示最佳可用权重

# 2. 使用预训练权重设置 ViT 模型实例
pretrained_vit = torchvision.models.vit_b_16(weights=pretrained_vit_weights).to(device)

# 3. 冻结基础模型的参数
for parameter in pretrained_vit.parameters():
    parameter.requires_grad = False

# 4. 更改分类器 head（设置随机种子以确保线性分类器的初始化一致）
set_seeds()
pretrained_vit.heads = nn.Linear(in_features=768, out_features=len(class_names)).to(device)
# pretrained_vit # 取消注释以查看模型输出
````

我们可以使用 Python 的 [`pathlib.Path().stat()`](https://docs.python.org/3/library/pathlib.html#pathlib.Path.stat) 方法的 `st_size` 属性来检查模型的大小（以字节为单位），并将模型的文件路径传递给它。

然后，我们可以将字节大小转换为兆字节。

````python
from pathlib import Path

# 获取模型的字节大小，然后转换为兆字节
pretrained_vit_model_size = Path("models/08_pretrained_vit_feature_extractor_pizza_steak_sushi.pth").stat().st_size // (1024*1024) # division converts bytes to megabytes (roughly)
print(f"Pretrained ViT feature extractor model size: {pretrained_vit_model_size} MB")
````

嗯，看起来我们为 Food Vision Mini 创建的 ViT 特征提取器模型的大小大约是 327 MB。

与 [07. PyTorch Experiment Tracking 第 9 节](https://www.learnpytorch.io/07_pytorch_experiment_tracking/#9-load-in-the-best-model-and-make-predictions-with-it) 中的 EffNetB2 特征提取器模型相比，怎么样呢？

| **模型** | **模型大小 (MB)** | **测试损失** | **测试准确率** |
| ----- | ----- | ----- | ------ |
| EffNetB2 特征提取器^ | 29 | ~0.3906 | ~0.9384 |
| ViT 特征提取器 | 327 | ~0.1084 | ~0.9384 |

> **注意：** ^ 参考中的 EffNetB2 模型使用了 20% 的披萨、牛排和寿司数据（图像量是原来的两倍），而 ViT 特征提取器则使用了 10% 的披萨、牛排和寿司数据。一个练习可以是用相同数量的数据训练 ViT 特征提取器模型，看看结果有多大改善。

EffNetB2 模型比 ViT 模型小约 11 倍，并且在测试损失和准确率上有相似的结果。

然而，当使用相同的数据（20% 的披萨、牛排和寿司数据）训练时，ViT 模型的结果可能会进一步提高。

但从部署的角度来看，如果我们比较这两个模型，我们需要考虑的是，ViT 模型的额外准确性是否值得大约 11 倍的模型大小增加？

也许这样一个大模型加载和运行的时间更长，且可能无法提供与 EffNetB2 相同的良好体验，后者虽然在性能上相似，但模型大小大大减少。
