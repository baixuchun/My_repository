```python
import matplotlib.pyplot as plt
import numpy as np

def beautify_line_chart():
    # 美化直线图
    plt.xlabel('x')
    plt.ylabel('f(x)')
    #确定坐标范围
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    #确定坐标刻度
    plt.xticks([-2,-1,0,1,2])
    plt.yticks([-2,-1,0,1,2])
    #确定坐标网格
    plt.grid(True)
    #横纵坐标统一标尺
    plt.gca().set_aspect('equal',adjustable='box')
    #画出图形
    plt.show()

x_array=np.linspace(-2,2,101)
fig, ax=plt.subplots(figsize = (4,4))

y_array=x_array
plt.plot(x_array,y_array)
beautify_line_chart()

```

```python
#绘制抛物线
x_array2=np.linspace(-2,2,101)
y_array2=x_array2**2-2
fig,ax=plt.subplots(figsize=(4,4))
plt.plot(x_array2,y_array2)
beautify_line_chart()
```

```python
#局部变量
global_x=8
print('global_x:',global_x)

def my_function(local_x):
    print('local_x:',local_x)
    
    global_x=88 #这里是局部变量
    print('global_x:',global_x)

my_function(9)
print('global:',global_x)
global global_x

#强调函数内某个变量可以将其变为全局变量
global_x=8
print('global_x:',global_x)

def my_function(local_x):
    print('local_x:',local_x)
    
    global global_x #强调全局变量
    global_x=88 
    print('global_x:',global_x)

my_function(9)
print('global:',global_x)
```

```python
#自定义函数
def matrix_multiplication(A,B):
    if len(A[0]) != len(B):
        raise ValueError('Error: check your matrix sizes')
    else:
        C=[[0] * len(B[0]) for _ in range(len(A))]
        for i in range(len(A)): #提取A的第i行
            for j in range(len(B[0])): #提取B的第i列，与A的行对应
                for k in range(len(A[0])):# 根据矩阵运算公式，计算C的第i行第i列的元素，一共需要相加k次（对应A的列数或B的行数）
                    C[i][j]+=A[i][k]*B[k][j]
        return C
A=[[1,0,-1,1],[0,-1,0,1]]
B=[[1,0],[0,-1],[-1,0],[1,1]]
c=matrix_multiplication(A,B)
print(c)
```

```python
#自定义函数产生全0函数
def creat_zero_matrix(rows,cols):
    matrix=[]
    a=[0]*cols
    for _ in range(rows):
        matrix.append(a)
    return matrix
creat_zero_matrix(9,8)
```

```python
#产生单位矩阵
def ifentity_matrix(size):
    matrix=[[0]*size for _ in range(size)]
    for i in range(size):
        matrix[i][i]=1
    return matrix
ifentity_matrix(9)

```

```python
#将向量展开为对角方阵
def diagonal_matrix(values):
    size=len(values)
    matrix=[[0]*size for _ in range(size)]
    for i in range(size):
        matrix[i][i]=values[i]
    return matrix
diagonal_matrix([1,2,3,4,5])
```

```python
#提取对角元素
def extract_main_diagonal(matrix):
    size=min(len(matrix),len(matrix[0]))
    new_matrix=[]
    for i in range(size):
        new_matrix.append(matrix[i][i])
    return new_matrix
extract_main_diagonal([[1,2,3],[12,3,34],[9,6,4]])
```

```python
# 计算方阵的迹
def trace(matrix):
    size=len(matrix)
    tr=0
    for i in range(size):
        tr+=matrix[i][i]
    return tr
trace([
    [1,2,3,4],
    [4,5,6,7],
    [1,3,5,6],
    [1,7,5,3]
])
```

```python
# 判断矩阵是否为对称矩阵
def is_symmetric(matrix):
    s='是'
    for m in range(len(matrix)):
        for n in range(len(matrix[0])):
            if matrix[m][n]!= matrix[n][m]:
                s='不是'
    return print(f'{matrix}{s}对称矩阵')

A=[
    [1,2,3],
    [2,4,4],
    [3,4,5]
]
B=[
    [1,2,3],
    [2,4,0],
    [0,5,6]
]
is_symmetric(A)
is_symmetric(B)
```

```python
#矩阵的转置
def transposed(matrix):
    new_matrix=[[0]* len(matrix) for i in range(len(matrix[0]))]
    for m in range(len(matrix[0])):
        for n in range(len(matrix)):
            new_matrix[m][n]=matrix[n][m]
    return new_matrix

B=[
    [1,2,3],
    [2,4,0],
    [0,5,6]
]
print(transposed(B))

''' 还可以用列表生成式来写一个转置'''

transpose=[[B[i][j] for i in range(len(B))] for j in range(len(B[0]))]
print(transpose)
```
